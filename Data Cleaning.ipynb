```python
# Importing libraries
import pandas as pd
import numpy as np
```

```python
#Importing dataset
df = pd.read_csv(r'C:\Users\Sirmoi Wekesa\Downloads\messy_synthetic_data.csv')
df.head()
```

```python
#Understanding structure
df.shape
df.columns
```

```python
# Cleaning column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')
```

```python
#Checking out column names again
df.columns
```

```python
# Checking missing values
df.isnull().sum()
```

```python
#Checking data types
df.dtypes
```

```python
#Changing to date time format
df['join_date'] = pd.to_datetime(df['join_date'], errors='coerce')
```

```python
df.dtypes
```

```python
#changing name column to string
df['name'] = df['name']. astype(str)
```

```python
df.dtypes
```

```python
#Handling missing values
df.isnull().sum()
```

```python
#replacing null values within the slary column with the mean
df['salary'] = df['salary'].fillna(df['salary'].mean())
```

```python
df.isnull().sum()
```

```python
df['age'] = df['age'].fillna(df['age'].median())
```

```python
df.isnull().sum()
```

```python
df['department'] = df['department'].fillna('unknown')
df['city'] = df['city'].fillna(df['city'].mode()[0])
df['join_date'] = df['join_date'].fillna('unknown')
df['email'] = df['email'].fillna('unknown')
```

```python
df.isnull().sum()
```

```python
# Checking duplicates
df.duplicated().sum()
```

```python
#View duplicated
df[df.duplicated()]
```

```python
#DROPPING DUPLICATES
df.drop_duplicates(inplace = True)
df.duplicated().sum()
```

```python
# checking data types
df.dtypes
```

```python
df['age'] = df['age'].astype(int)
df['join_date'] = pd.to_datetime(
    df['join_date'],
    errors='coerce',
    format='mixed'
)
```

```python
df.dtypes
```

```python
#Standardizing text
df['name'].unique()
```

```python
df['name'] = df['name'].replace({
    'Al!ce': 'Alice',
    'nan': 'NaN',
    'CHARLIE': 'Charlie',
    'bob': 'Bob'
})
    
```

```python
df['city'] = df['city'].str.title()
```

```python
df['city'].unique()
```

```python
df['department'].unique()
```

```python
df['department'] = df['department'].str.title()
```

```python
df['department'].unique()
```

```python
df['email'] = df['email'].str.lower()
```

```python
df['email'].unique()
```

```python
# Handling outliers
# Handling outliers
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['salary'] >= lower) & (df['salary'] <= upper)]
```

```python
# Fixing wrong entries
df[df['age'] < 0]

#Replace
df.loc[df['age'] < 0, 'age'] = np.nan
```
